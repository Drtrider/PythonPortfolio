# Creating a dataframe of stats for a given player

```python
import requests
import pandas as pd

player_id = 4361741
```

```python
def fetch_player_stats(player_id):
    """Fetches a players stats, and returns them as a json string"""
    url = f"http://sports.core.api.espn.com/v2/sports/football/leagues/nfl/athletes/{player_id}/statistics?lang=en&region=us"

    payload = {}
    headers = {}

    response = requests.request("GET", url, headers=headers, data=payload)

    return response.json()
```

```python
player_stats_response = fetch_player_stats(player_id)

print(player_stats_response)
```

```python
df_player_stats = pd.json_normalize(player_stats_response, record_path=['splits', 'categories', 'stats'], meta=['$ref', ['athlete', '$ref'], ['splits', 'id'], ['splits', 'name']], errors='ignore')
```
